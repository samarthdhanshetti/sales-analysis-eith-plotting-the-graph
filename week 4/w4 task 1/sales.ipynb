```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd .read_csv("retail_sales.csv")
df.head(10)
```

```python
df['OrderDate']
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Month'] = df['OrderDate'].dt.month
```

```python
df['OrderDate']
```

```python
df['revenue'] = df['UnitPrice'] * df['Quantity']
```

```python
df['revenue']
```

```python
total_revenue = df['revenue'].sum()
avg_Order_Value = df.groupby('OrderDate')['revenue'].sum().mean()
```

```python
print("Total Revenue:", total_revenue)
```

```python
print("Average Order Value:", avg_Order_Value)
```

```python
top_produts = df.groupby('Product')['revenue'].sum().sort_values(ascending = False)
```

```python
print("top product value",top_produts)
```

```python
top_produts.plot(kind='bar')
plt.title("top products of sales")
plt.xlabel('Product')
plt.ylabel("revenue")
plt.show()
```

```python
monthly_sales = df.groupby('Month')['revenue'].sum()
```

```python
monthly_sales.plot(marker ='o')
plt.title('mothly  sale of the product')
plt.xlabel('Month')
plt.ylabel('revenue')
plt.show()
```

```python
region_sales = df.groupby('Region')['revenue'].sum().sort_values(ascending =False)
print("region_sales",region_sales)
```

```python
region_sales.plot(kind='bar')
plt.title("region wise sale")
plt.xlabel('Region')
plt.ylabel('revenue')
plt.show()
```
